# Question 1

## 1.1

Empirical risk definition:

    R̂(w) = (1/I) ∑ᵢ L(y_i, d_i)

Substitute the squared error loss:

    R̂(w) = (1/I) ∑ᵢ₌₁ᴵ (y_i - d_i)²

Substitute the linear model y_i = w₀v_i + w₁h_i = wᵀx_i:

    R̂(w) = (1/I) ∑ᵢ₌₁ᴵ (wᵀx_i - d_i)²

Write as vector norm using matrix X = [x₁ᵀ; x₂ᵀ; ...; x_Iᵀ]:

    R̂(w) = (1/I)(Xw - d)ᵀ(Xw - d)
         = (1/I)‖Xw - d‖²

## 1.2

### 1.2.1 Gradient of the empirical risk

Empirical risk:

    R̂(w) = (1/I)‖Xw - d‖² 
         = (1/I)(Xw - d)ᵀ(Xw - d)

Expand:

    R̂(w) = (1/I)(wᵀXᵀXw - wᵀXᵀd - dᵀXw + dᵀd) 
         = (1/I)(wᵀXᵀXw - 2dᵀXw + dᵀd)

Gradient with respect to w:

    ∇R̂(w) = (1/I)(2XᵀXw - 2Xᵀd) 
          = (2/I)(XᵀXw - Xᵀd)

### 1.2.2 Gradient Descent Update Rule

    w(t+1) = w(t) - η∇R̂(w(t))
           = w(t) - (2η/I)(XᵀXw(t) - Xᵀd)


## 1.3 Minimizer of the empirical risk

Set gradient to zero:

    ∇R̂(w) = (2/I)(XᵀXw - Xᵀd) = 0

XᵀX is invertible, multiply both sides by (XᵀX)⁻¹:

    w* = (XᵀX)⁻¹Xᵀd

# 1.4

### 1.4.1 I = 2 (X ∈ R²ˣ²)

X is invertible, solve Xw = d directly:

    w* = X⁻¹d

Minimum empirical risk:

    R̂(w*) = (1/2)‖Xw* - d‖²
          = (1/2)‖XX⁻¹d - d‖²
          = (1/2)‖d - d‖²
          = 0

### 1.4.2 I ≥ 3 (X ∈ R²ˣ²)

I ≥ 3:

    -> X ∈ Rᴵˣ² is no longer square
    -> more equations than unknowns
    -> impossible to satisfy all Xw = d exactly
    -> R̂(w*) = (1/3)‖Xw* - d‖² > 0 (always have some non-zero error)

### 1.5

    w(1) = w* = (XᵀX)⁻¹Xᵀd

Substitute update rule:

    w(0) - (2η/I)(XᵀXw(0) - Xᵀd) = (XᵀX)⁻¹Xᵀd

Left side has two terms with w(0): w(0) and XᵀXw(0), Which scales differently with w(0)

-> No single η can make this equation hold for all starting points w(0)

# 1.6

Using matrix H = [η₁ 0; 0 η₂] for the update rule:

    w(1) = w(0) - H∇R̂(w(0))
         = w(0) - (2/I)H(XᵀXw(0) - Xᵀd)

One-Step Convergence

    w(0) - (2/I)H(XᵀXw(0) - Xᵀd) = (XᵀX)⁻¹Xᵀd

Choose H to satisfy:

    (2/I)HXᵀX = (XᵀX)⁻¹

which gives:

    η₁ = I/(2λ₁)
    η₂ = I/(2λ₂)
